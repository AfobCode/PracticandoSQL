## Finding Purchases problem in StrataScratch
Identify returning active users by finding users who made a second purchase within 7 days of any previous purchase

```python
SELECT user_id, created_at, 
    LAG(created_at) OVER (PARTITION BY user_id ORDER BY created_at), created_at - LAG(created_at) OVER (PARTITION BY user_id ORDER BY created_at) time_between_t
FROM amazon_transactions;
```

```python
-- After getting the difference between the previous transaction and the current one, now I can filter and get unique user_id 
WITH transaction_time AS (
    SELECT user_id, created_at, LAG(created_at) OVER (PARTITION BY user_id ORDER BY created_at), created_at - LAG(created_at) OVER (PARTITION BY user_id ORDER BY created_at) time_between_t
    FROM amazon_transactions
)
SELECT DISTINCT user_id
FROM transaction_time
WHERE time_between_t < 7;
```
